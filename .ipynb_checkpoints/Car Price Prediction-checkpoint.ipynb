### Cleaning the data

```python
df.duplicated().sum()
```

```python
# Dropping Duplicates
df.drop_duplicates(inplace=True)
df.shape
```

```python
# Fixing data format issues for Dorrs
df['Doors'].replace('04-May','4-5',inplace = True)
df['Doors'].replace('02-Mar','2-3',inplace = True)
df['Doors'].value_counts()
```

```python
df['Levy']
```

```python
# Replacing - with 0 
df['Levy'] = df['Levy'].apply(lambda x: int(x.replace('-','0')))
df['Levy']
```

```python
df['Engine volume']
```

```python
# Converting to numeric 
df['Engine volume'] = df['Engine volume'].apply(lambda x: float(x.replace('Turbo','')))
df['Engine volume']
```

```python
df['Mileage']
```

```python
# Converting to numeric 
df['Mileage'] = df['Mileage'].apply(lambda x: int(x.replace('km','')))
df['Mileage']
```

### Analysis by Manufacturer

```python
top_10_manu_by_models = clean_df.groupby(['Manufacturer']).Model.nunique().nlargest(10).reset_index()
```

```python
plt.figure(figsize=(15,5))
ax = sns.barplot(data = top_10_manu_by_models, x = 'Manufacturer', y = 'Model')
ax.bar_label(ax.containers[0]);
plt.title("Top 10 Manufacturers by number of models");
```

```python
top_10_manu = clean_df.Manufacturer.value_counts().head(10)
top_10_manu_data = clean_df[clean_df['Manufacturer'].isin(top_10_manu.index)]
top_10_manu_fuel_dist=top_10_manu_data.groupby(['Manufacturer','Fuel type'])['Fuel type'].count().unstack().fillna(0)
```

```python
top_10_manu_fuel_dist_perc = top_10_manu_fuel_dist.T.transform(lambda x: 100*x / x.sum())
top_10_manu_fuel_dist_perc
```

```python
top_10_manu_fuel_dist_perc.T.plot(kind='barh', stacked=True , figsize=(15,10));
plt.title("Distribution of fuel type by Manufacturer");
```

```python
top_10_manu_avg_mil=top_10_manu_data.groupby(['Manufacturer'])['Mileage'].mean()
ax = top_10_manu_avg_mil.to_frame().plot(kind='bar', stacked=True , figsize=(10,5));
mean_mileage = np.mean(top_10_manu_avg_mil, 0)
plt.axhline(mean_mileage, color='red', ls='dotted');
plt.title("Average Mileage for top 10 Manufacturers");
```

```python
top_10_manu_cat=top_10_manu_data.groupby(['Manufacturer','Category'])['Category'].count().unstack().fillna(0)
top_10_manu_cat_perc = top_10_manu_cat.T.transform(lambda x: 100*x / x.sum())
top_10_manu_cat_perc
```

```python
top_10_manu_cat_perc.T.plot(kind='bar', stacked=True , figsize=(15,10));
plt.title("Distribution of Category  by Manufacturer");
```

## Predicting Price

```python
non_numeric_cols = clean_df.select_dtypes(exclude=[np.number]).columns
encoded_df = clean_df.copy()
for cols in clean_df.columns:
    if cols in non_numeric_cols:
        encoded_df[cols] = LabelEncoder().fit_transform(clean_df[cols])
```

```python
encoded_df
```

```python
plt.figure(figsize = (15,15))
sns.heatmap(encoded_df.corr(),annot=True,center=0);
```

```python
#Helper function to create split
def test_train_split(df):    
    df_train = df.sample(frac=0.8,random_state=4)
    df_test = df.drop(df_train.index)
    return df_train,df_test 
```

```python
df_train,df_test  = test_train_split(encoded_df)
```

```python
print(f'Train: {df_train.shape} \nTest: {df_test.shape} ')
```

```python
#Method to Split X and Y
def split_X_Y(df):
    X = df.drop('Price',axis = 1)
    Y = df['Price']
    return X, Y
```

```python
X_train, y_train = split_X_Y(df_train)
X_test, y_test = split_X_Y(df_test)
```

```python
print(X_train.shape, y_train.shape )
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
rf = RandomForestRegressor(random_state=0)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
reg = GradientBoostingRegressor(random_state=0)
reg.fit(X_train, y_train)
reg.score(X_test, y_test)
```
